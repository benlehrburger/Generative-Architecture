```python
import os
import selenium
from selenium import webdriver 
import base64
import time
import urllib.request
```

```python
# open browser
DRIVER_PATH = '/usr/local/bin/chromedriver'
SAVE_FOLDER = './scrape'
GOOGLE_IMAGES = 'https://www.google.com/search?q=modern+architecture&tbm=isch&sxsrf=APwXEdcSmET7xcNhwbdjsgbXmKL0veRjnQ%3A1684346840342&source=hp&biw=771&bih=655&ei=2BdlZM-CEuucptQPnvapoAU&iflsig=AOEireoAAAAAZGUl6InrGXyRmGPWj_PjLAX9GeynSFDx&ved=0ahUKEwjPzuD4-Pz-AhVrjokEHR57ClQQ4dUDCAc&uact=5&oq=modern+architecture&gs_lcp=CgNpbWcQAzIECCMQJzIICAAQgAQQsQMyBQgAEIAEMgUIABCABDIFCAAQgAQyBQgAEIAEMgUIABCABDIFCAAQgAQyBQgAEIAEMgUIABCABDoICAAQsQMQgwFQAFiRK2DZMGgAcAB4AYAB_gGIAboRkgEGMTEuNC40mAEAoAEBqgELZ3dzLXdpei1pbWc&sclient=img'
driver = webdriver.Chrome(DRIVER_PATH)
driver.get(GOOGLE_IMAGES)
```

```python
# Scroll to the end of the page
def scroll_to_end():
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)
    print('scroll done')
```

```python
# execute scrape

counter = 0
for i in range(1, 10):
    
    scroll_to_end()
    image_elements = driver.find_elements_by_class_name('rg_i')
    print(len(image_elements))
    
    for image in image_elements: 
        
        if (image.get_attribute('src') is not None):
            
            my_image = image.get_attribute('src').split('data:image/jpeg;base64,')
            filename = SAVE_FOLDER + 'helmet'+str(counter)+'.jpeg'
            
            if(len(my_image) >1): 
                
                with open(filename, 'wb') as f: 
                    f.write(base64.b64decode(my_image[1]))
                    
            else: 
                print(image.get_attribute('src'))
                urllib.request.urlretrieve(image.get_attribute('src'), SAVE_FOLDER + 'helmet'+ str(counter)+'.jpeg')
            
            counter += 1
```
